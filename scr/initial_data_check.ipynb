# Initial Data Check

## Purpose

This notebook checks the raw 2024 HMDA dataset.

The goal is to understand the data before cleaning and modeling.

## Checks in This Notebook

This notebook checks:

- dataset shape
- first five rows
- column data types
- approve and deny sample counts

```python

import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# load raw data
data_path = Path("../data/raw/HMDA_2024.csv")
df_raw = pd.read_csv(data_path, low_memory=False)
```

```python
pd.set_option("display.max_rows", None)
```

```python
# Print data shape
print("Raw data shape:", df_raw.shape)
```

```python
# View first five rows
df_raw.head(3).T
```

```python
# View all column data types
dtype_summary = df_raw.dtypes.reset_index()
dtype_summary.columns = ["column_name", "data_type"]

dtype_summary
```

```python
# Check approve and deny sample counts

# HMDA action_taken:
# 1 = Loan originated
# 2 = Application approved but not accepted
# 3 = Application denied

approve_count = df_raw["action_taken"].isin([1, 2]).sum()
deny_count = (df_raw["action_taken"] == 3).sum()

approve_deny_summary = pd.DataFrame({
    "class": ["approve", "deny"],
    "count": [approve_count, deny_count]
})

approve_deny_summary["percentage"] = (
    approve_deny_summary["count"] / approve_deny_summary["count"].sum()
)

approve_deny_summary
```
